```python
import requests # from urllib.request import urlopen
from bs4 import BeautifulSoup

import pandas as pd
from datetime import datetime
import time
import re
```

```python
query = '일본차 할인'
url = 'https://search.daum.net/search?nil_suggest=btn&w=news&DA=SBC&cluster=y&q=%EC%9D%BC%EB%B3%B8%EC%B0%A8+%ED%95%A0%EC%9D%B8'

urls_list = []
```

```python
for i in range(5) : 
    url=url+"&p="+str(i)
    
    web = requests.get(url).content
    source = BeautifulSoup(web, 'html.parser')
   

    for urls in source.find_all('a', {'class' : "f_nb"}):
        if urls["href"].startswith("http://v.media.daum.net/v/"):
            urls_list.append(urls["href"])
        
```

```python
driver = webdriver.Chrome('C:\chromedriver.exe') 
comments=[]
for url in urls_list : 
    try:
        print(url+'하는중')
        driver.get(url)
        time.sleep(3)

        for i in range(0, 2) : # 더보기 개수가 충족되지 않으면 예외로넘어감 해결필요

              driver.find_element_by_class_name("link_fold").click()

              time.sleep(3)

              i +=1


        
        html = driver.page_source

        bs = BeautifulSoup(html, 'html.parser')

        contents = bs.find_all("p", {"class": "desc_txt" })

        for content in contents :
            comments.append(content.get_text())
            print(len(comments))
    except:
        print("왜안돼")

print(comments)
```

```python
# 유의미한 의미를 갖고 있지 않은 단어를 제외할 수 있습니다.
del_list = ['하다', '있다', '되다', '이다', '돼다', '않다', '그렇다', '아니다', '이렇다', '그렇다', '어떻다','\n','ㅋ','지금'
           ,'우리','여기','ㅋㅋㅋ','아직','같다','없다','오다','차다','보다','자다','아직도','들어오다','들이다','나다','살다'
           ,'넘다','들다','주다','다니다','답글','알다','세기','라이','현기','따라가다','해보다','잖다','많다','저렇게','그냥',
           '싶다','프로필','13','대표','계정','건가','아지','힘주다','어쩌구','ㅠㅠ','ㅉㅉ','ㅎㅎ','없다','위해','완전','절대','그게','아무'] 

word_cleaned = []
for word in raw_pos_tagged:
    if not word[1] in ["Josa", "Eomi", "Punctuation", "Foreign"]: # Foreign == ”, “ 와 같이 제외되어야할 항목들
        if (len(word[0]) != 1) & (word[0] not in del_list): # 한 글자로 이뤄진 단어들을 제외 & 원치 않는 단어들을 제외
            word_cleaned.append(word[0])
        
```
